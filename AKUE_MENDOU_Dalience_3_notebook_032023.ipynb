### <a id="C1">1. Le lien entre le genre d’un client et les catégories des livres achetés</a>

Les variables genre et catégorie de livre sont toutes les deux des variables qualitatives ordinales,
nous allons donc utiliser le test du Chi2.

> Pour **déterminer le lien entre ces deux varaibles catégorielles**, nous allons utiliser le **test de chi-carré** comme suit :
>- Le tableau de contingence affichera le nombre de clients pour chaque combinaison de genre et de catégorie de livre
>- le chi-carré et la p-value seront utilisés pour déterminer si la relation est significative,
>- et le coefficient de contingence de Cramer indiquera la force de la relation.

```python
# Tableau de contingence
table_cont = pd.crosstab(client_vente['sexe'], client_vente['categ'])

# Test du chi-carré
chi2, pval, dof, expected = chi2_contingency(table_cont)

# Coefficient de contingence de Cramer
n = table_cont.values.sum()
phi2 = chi2 / n
r, k = table_cont.shape
phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
rcorr = r - ((r-1)**2)/(n-1)
kcorr = k - ((k-1)**2)/(n-1)
cramer_v = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# Affichage des résultats
print("Tableau de contingence :")
display(table_cont)
print("Chi-carré : ", chi2)
print("P-value : ", pval)
print("Coefficient de contingence de Cramer : ", cramer_v)

# Affichage graphique
sns.set_style("whitegrid")
ax = sns.barplot(x="sexe", y="count", hue="categ", data=pd.melt(table_cont.reset_index(), id_vars=['sexe'], value_name='count'))
ax.set_title("Répartition des achats par genre et par catégorie")
ax.set_xlabel("Genre")
ax.set_ylabel("Nombre d'achats")
plt.show()
```

### Observations et recommandations: 
>-  On peut voir que la plupart des clients ont acheté des livres de la catégorie 0, suivi des livres de la catégorie 1 et 2.

>- Le **test de chi-carré** a une statistique de **147.119** et une **p-value très faible de 1.131e-32** (inférieure à 0,05), ce qui suggère qu'il y a **une relation significative entre le genre des clients et les catégories de livres achetées.** En d'autres termes, le genre des clients peut jouer un rôle important dans le choix des catégories de livres achetés. 

>- Le **coefficient de contingence de Cramer de 0.0146** suggère qu'il y a une faible association entre le genre et les catégories de livres achetées. Cela signifie que le **genre des clients n'a qu'un impact limité sur les types de livres achetés** dans cette librairie.

>**En conclusion**, il y a **une corrélation significative mais faible** entre le genre des clients et les catégories de livres achetés. Il est donc possible qu'il y ait d'autres facteurs qui influencent le choix d'achat de livres des clients, tels que l'âge, le niveau d'études, le revenu, les intérêts personnels, etc.

### <a id="C2.1">2.1. Le lien entre l’âge des clients et le montant total des achats</a>

```python
# Vérifier la distribution de l'âge et du Chiffre d'affaires

age = client_vente['age']
chiffre_affaires = client_vente['chiffre_affaires']

# Afficher les deux graphiques dans des sous-graphiques
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))

sns.histplot(age, kde=True, color="blue", ax=ax1)
ax1.set_title("Distribution de l'âge des clients")
ax1.set_xlabel("Âge")
ax1.set_ylabel("Nombre de clients")

sns.histplot(chiffre_affaires, kde=True, color="green", ax=ax2)
ax2.set_title("Distribution du chiffre d'affaires des ventes")
ax2.set_xlabel("Chiffre d'affaires")
ax2.set_ylabel("Nombre de ventes")

plt.tight_layout()
plt.show()
```

> Les variables 'age' et 'chiffre_affaires' ne suivent pas une distribution normale mais asymétrique ves la droite. Nous choisissons le **test de correlation non paramétrique de Spearman pour déterminer si la relation est significative ou non.**

```python
# Effectuer le test de corrélation non paramétrique de Spearman

corr, p_value = stats.spearmanr(client_vente['age'], client_vente['chiffre_affaires'])
print("Coefficient de corrélation de rang de Spearman = %.3f, p-value = %.3f" % (corr, p_value))

# Visualiser la correlation entre l'âge et le montant total des achats

plt.scatter(client_vente['age'], client_vente['chiffre_affaires'])

 # Ajouter la ligne de tendance
z = np.polyfit(client_vente['age'], client_vente['chiffre_affaires'], 1)
p = np.poly1d(z)
plt.plot(client_vente['age'], p(client_vente['age']), "r--")

 # Ajouter des labels pour les axes
plt.xlabel('Âge')
plt.ylabel('Chiffre d\'affaires')

 # Afficher le graphique
plt.show()
```

### Observations :  
>- Selon le test de Spearman, il y a **une corrélation négative faible mais significative** entre l'âge des clients et le montant total des achats. Cela suggère que les **clients plus jeunes ont tendance à dépenser plus que les clients plus âgés**.
### Recommandations:

>- La librairie devrait offrir des **rabais ou des promotions spéciales pour les clients plus âgés**, afin de les encourager à acheter plus.
>- La librairie devrait suivre les **tendances du marché**, en particulier en ce qui concerne les **nouveaux produits et les évolutions technologiques** telles que les **livres numériques et les audiobooks**, pour rester compétitive.

### <a id="C2.2">2.2. Le lien entre l’âge des clients et la fréquence d’achats</a>

```python
# Calcul de la fréquence d'achats par âge
freq_achats_age = client_vente.groupby('age')['quantite'].sum() / client_vente.groupby('age')['ref_client'].nunique()

# Création d'un dataframe pour le test de corrélation
data_corr = pd.DataFrame({'age': freq_achats_age.index, 'freq_achats': freq_achats_age.values})

# Test de corrélation de Spearman
corr, p_value = stats.spearmanr(data_corr['age'], data_corr['freq_achats'])

print("Coefficient de corrélation de rang de Spearman =", corr, ", p-value =", p_value)


# Visualisation de la relation entre l'âge et la fréquence d'achats
sns.scatterplot(x="age", y="freq_achats", data=data_corr)
plt.title("Relation entre l'âge et la fréquence d'achats")
plt.xlabel("Âge")
plt.ylabel("Fréquence d'achats")
plt.show()
```

### Observations et recommandations: 

> Le **coefficient de corrélation de rang de Spearman est de -0.076**, ce qui indique une **corrélation négative très faible** entre l'âge des clients et la fréquence d'achats. La **p-value est de 0.517**, ce qui est **supérieur à la valeur seuil de 0.05**, ce qui signifie qu'il n'y a **pas de preuve suffisante pour rejeter l'hypothèse nulle** selon laquelle il n'y a pas de corrélation significative entre l'âge des clients et la fréquence d'achats.

>- Le **coefficient de corrélation de Spearman est négatif**, ce qui indique une **relation inverse** entre l'âge des clients et la fréquence d'achat. Cependant, **la p-value est supérieure à 0,05**, ce qui suggère que la **corrélation n'est pas statistiquement significative** et qu'il n'y a donc pas de relation linéaire forte entre ces deux variables. 

>En conclusion, **l'âge des clients ne peut pas être considéré comme un facteur clé pour déterminer la fréquence d'achat.** D'autres facteurs peuvent avoir une influence plus importante sur la fréquence d'achat des clients.

### <a id="C2.3">2.3. Le lien entre l'âge des clients et la taille du panier moyen</a>

```python
# Regrouper les ventes par client et calculer la moyenne du chiffre d'affaires 
panier_moyen = client_vente.groupby('ref_client')['chiffre_affaires'].mean() 

# Convertir le résultat en DataFrame 
panier_moyen = panier_moyen.reset_index() 
panier_moyen.columns = ['ref_client', 'panier_moyen'] 

# Jointure des données de l'âge et du panier moyen par client 
age_panier = pd.merge(client_vente[['ref_client', 'age'] ], panier_moyen, on='ref_client') 

# Calculer le coefficient de corrélation de rang de Spearman 
corr_coef, p_value = stats.spearmanr(age_panier['age'], age_panier['panier_moyen']) 
print("Coefficient de corrélation de rang de Spearman = %.3f, p-value = %.3f" % (corr_coef, p_value)) 

# Visualisation de la corrélation 
sns.scatterplot(x='age', y='panier_moyen', data=age_panier) 
plt.title('Corrélation du rang de Spearman entre l\'âge des clients et le panier moyen') 
plt.show()
```

### Observations et recommandations : 

>- le **coefficient de corrélation de Spearman est positif**, ce qui signifie que l'âge et la taille du panier moyen ont une relation directe : **plus l'âge des clients est élevé, plus leur panier moyen est important.**

>- La **p-value est très faible (inférieure à 0,05)**, ce qui signifie que cette **corrélation est statistiquement significative**. Autrement dit, il y a de fortes chances que cette corrélation ne soit pas due au hasard.

>En conclusion, Cette corrélation positive suggère que **les clients plus âgés ont tendance à acheter plus de produits et à dépenser plus lors de leurs achats**. Ce qui peut être dû à la proportion des outliers dans les achats des clients plus âgés, car ils ont  une plus grande capacité financière.

### <a id="C2.4">2.4. Le lien entre l’âge des clients et les catégories des livres achetés</a>

> Pour savoir s'il existe un lien entre l'âge des clients et les catégories de livres achetés, nous allons utiliser **le test non paramétrique de Kruskal-Wallis** qui permet de tester l'hypothèse nulle selon laquelle les échantillons de différentes catégories ont été tirés d'une même distribution.

```python
# Sélectionner les variables utiles
data_age_categ = client_vente[['age', 'categ']]

# Créer des groupes en fonction des catégories de livres achetés
groupes = []
for categ in data_age_categ['categ'].unique():
    groupes.append(data_age_categ[data_age_categ['categ'] == categ]['age'])

# Effectuer le test de Kruskal-Wallis
stat, pval = stats.kruskal(*groupes)
print('Statistique de test:', stat)
print('p-value:', pval)

# Visualiser la correlation
# Créer la figure et les axes
fig, ax = plt.subplots(figsize=(10,6))
sns.boxplot(x='categ', y='age', data=data_age_categ)
ax.set_title('Comparaison des âges par catégorie')

# Ajouter des légendes et un titre
ax.set_xticklabels(['Catégorie 0', 'Catégorie 1', 'Catégorie 2'])
ax.set_ylabel('Âge')
ax.set_title('Comparaison des âges par catégorie')

plt.show()
```

### Observations: 
 
>- La **p-value** étant **inférieure** au seuil de **0,05**, on peut **rejeter l'hypothèse nulle** selon laquelle les âges des clients sont indépendants des catégories de livres achetées.

Le **test de Kruskal-Wallis nous permet de conclure que l'âge des clients est significativement lié aux catégories de livres achetées** dans la librairie.
